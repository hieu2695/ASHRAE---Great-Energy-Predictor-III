# Introduction


<div style="text-align: justify">
A revenue forecasting is the estimate of upcoming revenue of a company in a future period of time using the recorded history of transactions, sales. Projecting revenue not only helps managers and entrepreneurs to be aware of how much money they will earn from their operations but also produces a variety of business benefits such as budgetting expenses, managing the cash flow, justifying long-term hiring and investing decisions, executing strategic plans and scheduling production. However, the most crucial advantage of estimating revenues is to improve the marketing business to two primary targets: investors and customers. A well-researched and specific analysis including market trends and sales analysis can help to convince investors and lenders that the business is potential and lucrative. Analyzing customer patterns and transaction data is necessary for predicting customers' behavior and future sales which greatly contribute to product promotional strategies and a company's bottom line.</div>


<div style="text-align: justify">
<br/>
However, it is challenging task for marketing teams to analyze the transaction data and make predictions for future revenues. The first difficulty is the Pareto principle (also known as the 80/20 rule, the law of the vital few, or the principle of factor sparsity) which has been proven to be true for many businesses - 80% of sales come from 20% of clients. To be more specifict, the majority of transactions bring no (no transaction is executed) or inconsiderable amount of money while only a small percentage of customers contribute to the revenue. This rule results into the asymmetry and right skewness of transaction revenues, which negatively influences the forecasting. Another challenge is the dimension and cardinality of the transaction dataset. There is a variety of predictors to the revenue such as customer's ID, region, country, city, local time, operating system, web-browser,... Handling multiple features with numerous distinct values requires a lot of effort and precise in feature engineering.</div>


Our project works on the competition [Google Analytics Customer Revenue Prediction](https://www.kaggle.com/c/ga-customer-revenue-prediction/overview) to  analyze the Google Merchandise Store data and make predictions for revenue per customer in upcoming quarters. In this project, we will propose a feature engineering process to deal with high dimensional data and high-cardinality categorical features and construct 3 models including Linear Regression, Light Gradient Boosting Machine and Multiplier Perceptron to forecast the revenues.

## Load data

We need to load data and flatten the json fields in some features. These json fields will be transformed to columns in the dataframe.

```python
def load_json(csv_path, nrows=None):
    JSON_COLUMNS = ['device', 'geoNetwork', 'totals', 'trafficSource'] # columns that have json format
    
    df = pd.read_csv(csv_path, 
                     converters={column: json.loads for column in JSON_COLUMNS}, # load as json format
                     dtype={'fullVisitorId': 'str'},
                     nrows=nrows)
    
    for column in JSON_COLUMNS: # save json elements as columns
        column_as_df = json_normalize(df[column])
        column_as_df.columns = [f"{column}.{subcolumn}" for subcolumn in column_as_df.columns]
        df = df.drop(column, axis=1).merge(column_as_df, right_index=True, left_index=True)
    return df
```

## Handling Identifiers

Identifiers are feautures which receive unique values and make no contribution to the prediction. We can remove them to tidy our data.

```python
# combine training, validation and testing 
df = pd.concat([df_train, df_valid, df_test], sort=False) 
```

```python
# id_checker return identifiers (with unique values) in the data
def id_checker(df):
    """
    The identifier checker

    Parameters
    ----------
    df : dataframe
    
    Returns
    ----------
    The dataframe of identifiers
    """
    
    # Get the identifiers
    df_id = df[[var for var in df.columns 
                if df[var].nunique(dropna=True) == df[var].notnull().sum()]]
    
    return df_id
```

## Handling uncommon variables

There are features that are not in the training set but in the testing set and vice versa. These features are considered as uncommon variables and need to be removed from the data.

### Identifying common variables

```python
# common_var_checker identifies variables in both training set and testing set apart from the target

def common_var_checker(df_train, df_test, target):
    """
    The common variables checker

    Parameters
    ----------
    df_train : dataframe of training data
    df_test : dataframe of testing data
    target : the name of the target

    Returns
    ----------
    The dataframe of common variables between the training and testing data
    """
    
    # get the intersection of the columns of training set and testing set including the target
    df_common_var = pd.DataFrame(np.intersect1d(df_train.columns, np.union1d(df_test.columns, [target])),
                                 columns=['common var'])
                
    return df_common_var
```

## Handling date time variables

The "date" data is recorded as integers combining year, month and day. We need to separate the column into 3 specific columns including year, month and day.

```python
import datetime

def datetime_transformer(df, datetime_vars):
    """
    The datetime transformer

    Parameters
    ----------
    df : dataframe
    datetime_vars : the datetime variables
    
    Returns
    ----------
    The dataframe where datetime_vars are transformed into the following 3 datetime types:
    year, month, day
    
    """
    # The dictionary with key as datetime type and value as datetime type operator
    dict_ = {'year'   : lambda x : x.dt.year,
             'month'  : lambda x : x.dt.month,
             'day'    : lambda x : x.dt.day }
    
    # make a copy of df that will contain 3 datetime variables
    df_datetime = df.copy(deep=True)
    
    # for each column that is datetime variable
    for var in datetime_vars:
        # cast the integer data to datetime 
        df_datetime[var] = df_datetime[var].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
        
        
        # for each item (datetime_type and datetime_type_operator) in dict_
        for datetime_type, datetime_type_operator in dict_.items():
            # add a new variable to df_datetime where:
            # the variable's name is var + '_' + datetime_type
            # the variable's values are the ones obtained by datetime_type_operator
            df_datetime[var + '_' + datetime_type] = datetime_type_operator(df_datetime[var])
            
    # remove datetime_vars from df_datetime
    df_datetime = df_datetime.drop(columns=datetime_vars)
                
    return df_datetime
```

```python
def nan_checker(df):
    """
    The NaN checker

    Parameters
    ----------
    df : dataframe
    
    Returns
    ----------
    The dataframe of variables with NaN, their proportion of NaN and dtype
    """
    
    # get the variables with NaN, their proportion of NaN and dtype
    df_nan = pd.DataFrame([[var, df[var].isna().sum() / df.shape[0], df[var].dtype]
                           for var in df.columns if df[var].isna().sum() > 0],
                          columns=['var', 'proportion', 'dtype'])
    
    # sort df_nan in accending order of the proportion of NaN
    df_nan = df_nan.sort_values(by='proportion', ascending=False)
    
    return df_nan
```

```python
# cat_var_checker identifies categorical variables in the data
def cat_var_checker(df):
    """
    The categorical variable checker

    Parameters
    ----------
    df: the dataframe
    
    Returns
    ----------
    The dataframe of categorical variables, their number of unique value, the most frequent value
    and its frequency as percentage
    """
    
    length = len(df)
    
    # get the dataframe of categorical variables and their number of unique value
    # set dropna as False to consider null as a categorical value
    df_cat = pd.DataFrame([[var, df[var].nunique(dropna=False), Counter(df[var]).most_common(1)[0][0],
                            Counter(df[var]).most_common(1)[0][1]/length ] 
                           for var in df.columns if df[var].dtype == 'object'],
                          columns=['var', 'nunique','most_frequent','%frequency'])
    
    # sort df_cat in accending order of the number of unique value
    df_cat = df_cat.sort_values(by='nunique', ascending=False)
    
    return df_cat
```

### Selecting encoding techniques

```python
df_cat_not_rm.reset_index(drop=True) # print df_cat_not_rm
```

There are many categorical variables that have high number of unique values, which is the high cardinality problem. To solve this issue in our project, we will use hashing-encoding technique to encode variables with the number of unique values more than 50. The remaining variables will be encoded by one-hot-encoding method.

```python
# get the dataframe of variables that will be hashing encoded
cat_hash = df_cat_not_rm[df_cat_not_rm["nunique"] > 50]

# print cat_hash
cat_hash.reset_index(drop=True)
```

```python
# get the dataframe of variables that will be hashing encoded
cat_onehot = df_cat_not_rm[df_cat_not_rm["nunique"] < 51]

# print cat_hash
cat_onehot.reset_index(drop=True)
```

### One-hot encoding categorical features

Categorical features will be one-hot encoded using pd.get_dummies().

```python
# one-hot-encode the categorical features in the combined data
df = pd.get_dummies(df, columns=np.setdiff1d(np.intersect1d(df.columns, cat_onehot['var']), [target]))

# print the first 5 rows of df
df.head()
```

### Hashing encoding categorical features

To perform hashing encoding, we will use HashingEncoder function from category_encoders package.

```python
import category_encoders as ce 
df_index = df.index # save the index of df
df_enc = df.copy(deep=True) # get a copy of df
```

```python
# this cell presents the hashing encoding task which may take a lof time to finish

for var in cat_hash['var']:
    var_enc = df_enc[[var]] # a dataframe containing the categorical that will be encoded
    ce_hash = ce.HashingEncoder(cols = [var]) # define the HashingEncoder method for var
    var_enc = ce_hash.fit_transform(var_enc) # fit HashingEncoder to the data and transform it
    
    # we use the default n_components = 8 so var_enc will have 8 columns col_0, col_1, ..., col_7
    
    var_enc.set_index(df_index, inplace=True) # set the index of the encoded dataframe as df
    df_enc = pd.concat([df_enc , var_enc], axis=1) # combine df and var_enc horizontally
    
    # rename col_ columns using the name of the variable
    df_enc = df_enc.rename({'col_0': var + "_" + "0", 
                    'col_1': var + "_" + "1",
                    'col_2': var + "_" + "2",
                    'col_3': var + "_" + "3",
                    'col_4': var + "_" + "4",
                    'col_5': var + "_" + "5",
                    'col_6': var + "_" + "6",
                    'col_7': var + "_" + "7"}, axis=1)


df_enc = df_enc.drop(columns = cat_hash["var"]) # drop the original categorical variables that were hashing encoded
    
```

## Hyperparameter tuning and model selection

### Getting the predefined split cross-validator

```python
from sklearn.model_selection import PredefinedSplit

# Combine the feature matrix in the training and validation data
X_train_valid = pd.concat([X_train, X_valid], ignore_index=True, sort=False)

# Combine the target vector in the training and validation data
y_train_valid = pd.concat([y_train, y_valid], ignore_index=True, sort=False)

# Get the indices of training and validation data
train_valid_idxs = np.append(np.full(X_train.shape[0], -1), np.full(X_valid.shape[0], 0))

# The PredefinedSplit
ps = PredefinedSplit(train_valid_idxs)
```

```python
param_grids = {}
```

### The parameter grid for LinearRegression


Since the sklearn LinearRegression is based on the normal equation (a closed-form solution), it is not essential to fine-tune the hyperparameters of the model.

```python
param_grids['lr'] = [{}]
```

### The parameter grid for LightGBM


The hyperparameters we want to fine-tune are:
- learning_rate 
- min_data_in_leaf
- reg_alpha
- reg_lambda

```python
# The grids for learning_rate
learning_rate_grids = [10 ** i for i in range(-5, 0)]

# The grids for min_data_in_leaf, we set high values for big dataset to reduce overfitting
min_data_in_leaf_grids = [500, 5000]

# The grids for reg_alpha
reg_alpha_grids = [10 ** i for i in range(-3, 2)]

# The grids for reg_lambda
reg_lambda_grids = [10 ** i for i in range(-3, 2)]

# Update param_grids
param_grids['lgbm'] = [{'model__learning_rate': learning_rate_grids,
                        'model__min_data_in_leaf': min_data_in_leaf_grids,
                        'model__reg_alpha': reg_alpha_grids,
                        'model__reg_lambda': reg_lambda_grids }]
```

```python
# The grids for alpha
alpha_grids = [10 ** i for i in range(-6, -1)]

# The grids for learning_rate_init
learning_rate_init_grids = [10 ** i for i in range(-5, 0)]

# Update param_grids
param_grids['mlpr'] = [{'model__alpha': alpha_grids,
                        'model__learning_rate_init': learning_rate_init_grids}]
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# The list of [best_score_, best_params_, best_estimator_] obtained by GridSearchCV
best_score_param_estimator_gs = []

for acronym in pipes.keys():
    # GridSearchCV
    gs = GridSearchCV(estimator=pipes[acronym],
                      param_grid=param_grids[acronym],
                      scoring='neg_mean_squared_error',
                      n_jobs=2,
                      cv=ps,
                      return_train_score=True)
        
    # Fit the pipeline
    gs = gs.fit(X_train_valid, y_train_valid)
    
    # Update best_score_param_estimator_gs
    best_score_param_estimator_gs.append([gs.best_score_, gs.best_params_, gs.best_estimator_])
    
    # Sort cv_results in ascending order of 'rank_test_score' and 'std_test_score'
    cv_results = pd.DataFrame.from_dict(gs.cv_results_).sort_values(by=['rank_test_score', 'std_test_score'])
    
    # Get the important columns in cv_results
    important_columns = ['rank_test_score',
                         'mean_test_score', 
                         'std_test_score', 
                         'mean_train_score', 
                         'std_train_score',
                         'mean_fit_time', 
                         'std_fit_time',                        
                         'mean_score_time', 
                         'std_score_time']
    
    # Move the important columns ahead
    cv_results = cv_results[important_columns + sorted(list(set(cv_results.columns) - set(important_columns)))]

    # Write cv_results file
    cv_results.to_csv(path_or_buf='./cv_results/GridSearchCV/' + acronym + '.csv', index=False)
```

## Kaggle Leaderboard score

We will use three best models from GridSearchCV and calculating the competition leaderboard score.

```python
from sklearn import metrics
def LBscore(model): # this function use the input model and 
                    # return the kaggle leaderboard score and the prediction dataframe
    
    
    valid = valid_ids.copy(deep=True) # create a copy of valid_ids
    valid.reset_index(drop=True, inplace=True) # reset its index
    
    y_valid_pred = model.predict(X_valid) # predict log revenues in validation set
    

    # Get the dataframe of y_valid_pred, which has the same shape as df_train
    df_y_valid_pred = pd.DataFrame(np.tile(y_valid_pred.reshape(-1, 1), df_train.shape[1]),
                                   columns=df_train.columns)
    

    # reverse transform from the standardized scale
    df_y_valid_pred = pd.DataFrame(ss.inverse_transform(df_y_valid_pred),
                              columns=df_train.columns)
    
    # due to the predictions of log have some negative values
    # we need to convert them to 0s
    y_valid_pred = df_y_valid_pred[target].copy(deep=True) # create a copy of non-standardized predicted values
    y_valid_pred[y_valid_pred < 0] = 0 # replace negative values as 0
    df_y_valid_pred[target] = y_valid_pred  # apply new predicted values
    
    # transform from the natural logarithm scale to original scale
    df_y_valid_pred[target] = np.expm1(df_y_valid_pred[target])
    
    valid["PredictedRevenue"] = df_y_valid_pred[target] # add predicted values to dataframe valid 
    
    # calculate the total transaction per user
    valid = valid.groupby("fullVisitorId")[target, "PredictedRevenue"].sum().reset_index()
    valid[target] = np.log1p(valid[target]) # transform the observed values to log scale
    valid["PredictedRevenue"] = np.log1p(valid["PredictedRevenue"]) # transform the predicted values to log scale
    
            
    # calculate the score using the competition metric
    score = np.sqrt(metrics.mean_squared_error(valid[target].values, valid["PredictedRevenue"].values))
    
    return valid, score
```

- In the encoding process, we can vary the number of components in HashingEncoder to obtain better encoding results. Moreover, there are plenty of encoding techniques such as Binary encoder, Contrast encoder and Bayesian encoder that would be more efficient for this data.
